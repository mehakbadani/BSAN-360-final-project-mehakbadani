## Part 2: Pivot Tables (1.5 Points)

Pivot tables provide a multi-dimensional summary of your data, allowing you to calculate metrics across two or more categorical axes.

**Your Task:**
1.  Construct a pivot table using `pd.pivot_table()`.
2.  Define the `index`, `columns`, and `values` parameters.
3.  If your dataset does not have suitable dimensions for a pivot table, explain the omission below.

```python
df_pivot = pd.pivot_table(
    df,
    values='price',
    index='neighbourhood_group',
    columns='room_type',
    aggfunc='mean'
)

print(df_pivot)
```

**Reflection & Justification:**
In Part 2, I created a pivot table using neighbourhood_group as the index and room_type as the columns, while calculating the average price of listings. This structure allows us to compare how listing prices vary across both boroughs and room types at the same time.

The pivot table shows that entire homes/apartments generally have the highest average prices in every borough, while private rooms and shared rooms are cheaper. Manhattan has the highest average prices across most room types, while boroughs like the Bronx and Queens tend to have lower average prices.

This table highlights how both location and room type influence Airbnb listing prices and provides a clear summary of pricing trends across different segments of the dataset.

## Part 3: Cross-Tabulation (1.0 Point)

```python
df_cross = pd.crosstab(df['neighbourhood_group'], df['room_type'])

print(df_cross)
```

**Reflection & Justification:**

From the results, we can observe that Manhattan and Brooklyn have the highest number of listings overall. Entire homes and private rooms are much more common than shared rooms across all boroughs. This frequency distribution helps us understand how Airbnb listings are structured in different locations and highlights which types of accommodations are most common in each neighbourhood group.
